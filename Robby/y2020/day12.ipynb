```python
from autoloader import Autoloader
def get_2020_data(day):
    puz_data_url = f"https://adventofcode.com/2020/day/{day}/input"
    al = Autoloader(puz_data_url)
    al.connect()
    data = al.fetch()
    return data
```

```python
d12tdata1="""F10
N3
F7
R90
F11"""
```

```python
import numpy as np
class ship():
    def __init__(self):
        self.coords = np.zeros(2)
        self.facing = np.array((1.,0.))
        self.move_dict = {
            'N': np.array((0,1)),
            'S': np.array((0,-1)),
            'E': np.array((1,0)),
            'W': np.array((-1,0)),
            'F': self.facing
        }
        self.rot_dict = {
            'R': -1,
            'L': 1
        }
    def rotate(self, theta):
        theta = np.radians(theta)
        eps = np.finfo(theta.dtype).eps
        self.facing = np.array((
            (np.cos(theta), -np.sin(theta)),
            (np.sin(theta), np.cos(theta))
        )) @ self.facing
        self.move_dict['F'] = self.facing
    def move(self, act):
        if act[0] in self.move_dict:
            self.coords += self.move_dict[act[0]]*int(act[1:])
        elif act[0] in self.rot_dict:
            self.rotate(int(act[1:])*self.rot_dict[act[0]])



tester = ship()
eps = np.finfo(np.float32).eps
print(tester.coords)
print(tester.facing)
any(map(tester.move, d12tdata1.splitlines()))
print(tester.coords)
```

```python
import scipy.spatial.distance as spsd
d12 = ship()
any(map(d12.move, d12data.splitlines()))
print(np.sum(np.abs(d12.coords)))
eps = np.finfo(d12.coords.dtype).eps
print(eps)
```

```python
class waypoint_ship(ship):
    def __init__(self):
        super(waypoint_ship, self).__init__()
        self.facing = np.array((10,1))
    
    def move(self, act):
        if act[0] == 'F':
            self.coords += int(act[1:])*self.facing
        elif act[0] in self.move_dict:
            self.facing += self.move_dict[act[0]]*int(act[1:])
        elif act[0] in self.rot_dict:
            self.rotate(int(act[1:])*self.rot_dict[act[0]])

tester = waypoint_ship()

any(map(tester.move, d12tdata1.splitlines()))
print(np.sum(np.abs(tester.coords)))
```
